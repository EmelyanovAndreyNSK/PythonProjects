```
from tensorflow.keras.models import model_from_json #save load model

def SaveModel(model, path, name):
    with open(path + "/" + name + ".json", "w") as json_file:
        model_json = model.to_json()
        json_file.write(model_json)
    model.save_weights(path + "/" + name + ".h5")

def LoadModel(path, name):
    with open(path + "/" + name + ".json", "r") as json_file:
        loaded_model_json = json_file.read()
        model = model_from_json(loaded_model_json)
    model.load_weights(path + "/" + name + ".h5")
    return model
```

```
import math #batch manager
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import re
from tqdm import tqdm

def natural_sorting(a):
    a.sort(key=lambda var: [int(x) if x.isdigit() else x for x in re.findall(r'(-{0,1}[0-9]+)|[^0-9]|', var)])
    return a


def mask_image_to_hot_one(img):
    result = np.full((img.shape[0], img.shape[1], 3), 0)
    result[:,:,2] = 1 #[0,0,1] by default
    index_black = np.where((img < [100,100,100]).all(axis=2))
    index_white = np.where((img > [150,150,150]).all(axis=2))
    result[index_black] = [1,0,0]
    result[index_white] = [0,1,0]
    return result

def split_image(img, frame_size=32 * 8, border=32):
    core_size = frame_size - 2 * border
    assert core_size > 0

    new_shape = list(img.shape)
    new_shape[0] += frame_size
    new_shape[1] += frame_size
    extended_img = np.empty(new_shape, img.dtype)
    s = (img.shape[0], img.shape[1])
    extended_img[border:border + s[0], border:border + img.shape[1]] = img
    extended_img[border:border + s[0], 0:border] = extended_img[border:border + s[0], 2 * border:border:-1]
    extended_img[border:border + s[0], s[1] + border:] = extended_img[border:border + s[0], s[1] + border - 2:s[1] - core_size - 2:-1]
    extended_img[0:border, :] = extended_img[2 * border:border:-1, :]
    extended_img[s[0] + border:, :] = extended_img[s[0] + border - 2:s[0] - core_size - 2:-1, :]
    # plt.imshow(extended_img); plt.show()

    result = []
    for y in range(0, img.shape[0], core_size):
        for x in range(0, img.shape[1], core_size):
            result.append(extended_img[y:y + frame_size, x:x + frame_size])
    return result

def unsplit_image(array, output_shape, frame_size, border):
    core_size = frame_size - 2 * border
    assert core_size > 0

    new_shape = list(output_shape)
    new_shape[0] += core_size
    new_shape[1] += core_size
    extended = np.empty(new_shape, array[0].dtype)
    i = 0
    for y in range(0, output_shape[0], core_size):
        for x in range(0, output_shape[1], core_size):
            extended[y:y + core_size, x:x + core_size] = array[i][border:border + core_size, border:border + core_size]
            i += 1
    return extended[:output_shape[0], :output_shape[1]]

def unsplit_image_average(array, output_shape, frame_size, border):
    core_size = frame_size - 2 * border
    assert core_size > 0

    new_shape = list(output_shape)
    new_shape[0] += frame_size
    new_shape[1] += frame_size
    extended = np.empty(new_shape, np.float64)
    count = np.empty(new_shape, np.float64)
    i = 0
    for y in range(0, output_shape[0], core_size):
        for x in range(0, output_shape[1], core_size):
            extended[y:y + frame_size, x:x + frame_size] += array[i]
            count[y:y + frame_size, x:x + frame_size] += 1
            i += 1
    extended[count>0] /= count[count>0]
    return np.asarray(extended[border:border + output_shape[0], border:border + output_shape[1]], dtype=array[0].dtype)


def generate_x_file(path, file, IMG_SIZE, BORDER):
    try:
        img_array = cv2.imread(os.path.join(path, file), cv2.IMREAD_COLOR)
        assert img_array is not None
        # plt.imshow(img_array), plt.show()
        data = split_image(img_array, IMG_SIZE, BORDER)
        return data
    except OSError as e:
        print("OSError. Bad img most likely", e, os.path.join(path, file))



def generate_x_folder(path, IMG_SIZE, BORDER, img_postfix=".png"):
    training_data = []
    n = 0
    for img in tqdm(natural_sorting(os.listdir(path))):
        if img_postfix is not None and img.endswith(img_postfix):
            training_data += generate_x_file(path, img, IMG_SIZE, BORDER)
            n += 1
    return np.stack(training_data), n


def generate_xy_folder(path, IMG_SIZE, BORDER, img_postfix="original.png", mask_postfix="fimal_mask.png"):
    training_data = []
    training_mask = []
    n = 0
    for img in tqdm(natural_sorting(os.listdir(path))):
        if img.endswith(img_postfix):
            mask = img[:-len(img_postfix)] + mask_postfix
            training_data += generate_x_file(path, img, IMG_SIZE, BORDER)
            training_mask += generate_x_file(path, mask, IMG_SIZE, BORDER)
            n += 1
    return np.stack(training_data), np.stack(training_mask), n

def SaveBatch(x, y, path, name):
    np.savez_compressed(path + "/" + name, x=x, y=y)

def LoadBatch(path, name):
    data = np.load(path + "/" + name)
    return data['x'], data['y']
```

```
import SaveLoadModel #Test network
import BatchManager
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

def get_mask(arr):
    arr = np.argmax(arr, 2)
    pic = np.empty((arr.shape[0], arr.shape[1], 3), dtype=np.int)
    pic[arr == 0] = np.array([0, 0, 0])
    pic[arr == 1] = np.array([255, 255, 255])
    pic[arr == 2] = np.array([255, 0, 0])
    return pic

def get_circle(arr):
    inner = arr
    inner[inner != 1] = 0
    N, labels, stats, centroids = cv2.connectedComponentsWithStats(inner.astype(np.int8))
    result = []
    for i in range(1, N):
        result.append({
            'x': centroids[i][0],
            'y': centroids[i][1],
            'size': stats[i][4],
            'r': np.sqrt(stats[i][4] / math.pi)})
    return result

def main():
    DATADIR = "../HSV 53"
    OUTPUTDIR = "../output"
    OUTPUT_SHAPE = [184, 176, 3]
    IMG_SIZE = 32 * 6
    BORDER = 32 * 1
    x_train, img_n = BatchManager.generate_x_folder(DATADIR, IMG_SIZE=IMG_SIZE, BORDER=BORDER, img_postfix=".png")
    # x_train, y_train, img_n = BatchManager.generate_xy_folder(DATADIR, IMG_SIZE=IMG_SIZE, BORDER=BORDER, img_postfix="original.png", mask_postfix="fimal_mask.png")

    model = SaveLoadModel.LoadModel("../model", "5")

    # x_train = np.float32(x_train)
    predict = model.predict(x_train)


    active_bubbles = []
    history = []
    last_id = 0
    N = x_train.shape[0] // img_n
    for i in range(x_train.shape[0] // N):
        orig = BatchManager.unsplit_image(x_train[i*N:(i+1)*N], output_shape=OUTPUT_SHAPE, frame_size=IMG_SIZE, border=BORDER)
        mask = np.argmax(BatchManager.unsplit_image(predict[i*N:(i+1)*N], output_shape=OUTPUT_SHAPE, frame_size=IMG_SIZE, border=BORDER), 2)
        # true_mask = get_mask(BatchManager.unsplit_image(y_train[i*N:(i+1)*N], output_shape=[600, 800, 3], frame_size=IMG_SIZE, border=BORDER))
        # orig = x_train[i]
        # mask = get_mask(predict[i])
        # true_mask = get_mask(predict[i])
        orig[mask==1] = orig[mask==1] * [0.75,1,1] + [63, 0, 0]
        orig[mask==2] = [0, 0, 255]
        # for j in range(0, 600, IMG_SIZE-2*BORDER):    #split lines
        #     orig[j, :] = [0, 127, 0]
        #     orig[:, j] = [0, 127, 0]
        circles = get_circle(mask)
        for c in circles:
            orig = cv2.circle(orig, (int(c['x']), int(c['y'])), int(c['r']), [0, 255, 0], 1)
            orig[int(c['y']), int(c['x'])] = [0, 255, 0]

        cv2.imwrite(OUTPUTDIR + "/" + str(i) + ".png", orig)

        current_bubbles = []
        for c in circles:
            new = True
            for e in active_bubbles:
                if math.hypot(e['x'] - c['x'], e['y'] - c['y']) < 10:
                    new = False
                    current_bubbles.append({
                        'id': e['id'],
                        'age': e['age'] + 1,
                        'frame': i,
                        'x': c['x'],
                        'y': c['y'],
                        'size': c['size'],
                        'r': c['r']})
                    active_bubbles.remove(e)
                    break
            if new:
                current_bubbles.append({
                    'id': last_id,
                    'age': 0,
                    'frame': i,
                    'x': c['x'],
                    'y': c['y'],
                    'size': c['size'],
                    'r': c['r']})
                last_id += 1
        active_bubbles = current_bubbles
        history += active_bubbles
        # fig = plt.figure(figsize=(8, 8))
        # columns = 1
        # rows = 3
        # fig.add_subplot(rows, columns, 1)
        # plt.imshow(orig)
        # fig.add_subplot(rows, columns, 2)
        # plt.imshow(mask)
        # fig.add_subplot(rows, columns, 3)
        # plt.imshow(true_mask)
        # plt.show()
        # plt.savefig("C:/Users/Sasha/Documents/Bubbles/output/" + str(i))

    x_axis = 'frame'
    y_axis = 'r'
    for i in range(last_id):
        x = list(map(lambda x: x[x_axis], filter(lambda x: x['id'] == i, history)))
        y = list(map(lambda x: x[y_axis], filter(lambda x: x['id'] == i, history)))
        plt.plot(x, y, marker='.')
    # naming the x axis
    plt.xlabel(x_axis)
    # naming the y axis
    plt.ylabel(y_axis)
    # giving a title to my graph
    plt.title('History')
    plt.savefig(OUTPUTDIR + "/" + "Graph" + ".png", dpi=200)
    # function to show the plot
    plt.show()
    end = 1



if __name__ == "__main__":
    main()
```
