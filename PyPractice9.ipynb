```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import io
link = 'https://raw.githubusercontent.com/EmelyanovAndreyNSK/PythonTasks/master/company_sales_data.csv'
Data = pd.read_csv(io.StringIO(requests.get(link).text))
Data
```

```
plt.figure(figsize=(8, 6))
plt.plot(Data['month_number'], Data['total_profit'])
plt.title('Company profit per month')
plt.xlabel('Month Number'), plt.ylabel('Total profit')
plt.show()
```

```
plt.figure(figsize=(8, 6))
plt.plot(Data['month_number'], Data['total_profit'], linestyle='--', marker='o', color='r',  markerfacecolor='k', linewidth=3, label='Profit data of last year')
plt.title('Company sales data of last year')
plt.xlabel('Month Number'), plt.ylabel('Sold units number')
plt.legend()
plt.show()
```

```
labels = {'Face cream': 'facecream','Face wash': 'facewash','Tooth Paste': 'toothpaste','Bathing Soap': 'bathingsoap','Shampoo': 'shampoo', 'Moisturizer': 'moisturizer'}
plt.figure(figsize=(8, 6))
for label in labels:
    plt.plot(Data['month_number'], Data[labels[label]], marker='o', label=f'{label} Sales Data')
plt.title('Sales data')
plt.xlabel('Month Number'), plt.ylabel('Sales units in number')
plt.legend()
plt.show()
```

```
plt.figure(figsize=(8, 6))
plt.scatter(Data['month_number'], Data['toothpaste'], label='Tooth Paste Sales Data')
plt.grid(linestyle='--')
plt.title('Tooth Paste Sales Data')
plt.xlabel('Month Number'), plt.ylabel('Number of units sold')
plt.legend()
plt.show()
```

```
n = Data.plot(x='month_number', y=['facecream', 'facewash'], kind='bar', rot=0)
n.legend(['Face cream Sales Data', 'Face wash Sales Data'])
plt.title('Face wash and Face cream Sales Data')
plt.xlabel('Month Number'), plt.ylabel('Sales units in number')
n.grid(linestyle='--')
plt.show()
```

```
n = Data.plot(x='month_number', y='bathingsoap', kind='bar', rot=0, legend=False)
n.grid(linestyle='--')
plt.title('bathingsoap Sales Data')
plt.xlabel('Month Number'), plt.ylabel('Sales units in number')
plt.show()
```

```
plt.figure(figsize=(8, 6))
profit_range=[150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(Data['total_profit'], label='Profit data', bins=profit_range)
plt.title('Profit data')
plt.xlabel('Profit range in dollar'), plt.ylabel('Actual Profit in dollar')
plt.show()
```

```
labels = {'Face cream': 'facecream','Face wash': 'facewash','Tooth Paste': 'toothpaste','Bathing Soap': 'bathingsoap','Shampoo': 'shampoo', 'Moisturizer': 'moisturizer'}
sum_ = [sum(Data[labels[label]]) for label in labels]
plt.figure(figsize=(8, 6))
plt.pie(sum_,  labels=labels, autopct='%1.1f%%')
plt.title('Sales data')
plt.legend()
plt.show()
```

```
plt.subplot(211)
plt.plot(Data['month_number'], Data['bathingsoap'], marker='o',color='k')
plt.title('Sales data of Bathingsoap')
plt.xticks([])
plt.subplot(212)
plt.plot(Data['month_number'], Data['facewash'], marker='o', color='r')
plt.title('Sales data of a facewash')
plt.xlabel('Month Number'), plt.ylabel('Sold units in number')
plt.show()
```

```
n=Data.plot.area(y=labels.values(),stacked=True, figsize=(8,6))
n.legend([f'{label} Sales Data' for label in labels])
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number'), plt.ylabel('Sales units in number')
plt.show()
```
